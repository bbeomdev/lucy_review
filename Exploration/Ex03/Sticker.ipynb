```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
# 이미지 경로가 옳지 않다고 오류가 떠서 경로 확인
print(os.getcwd())
```

```python
# 준비한 이미지 읽고 출력하기
my_image_path = '/workspace/userdisk/AIFFEL_quest_rs/Exploration/Ex03/images/image.png'
img_bgr = cv2.imread(my_image_path)
img_show = img_bgr.copy()
plt.imshow(img_bgr)
plt.show()
```

```python
# opencv는 BGR을 사용하므로 RGB로 바꿔주기
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
# dlib를 활용해 hog detector 선언
detector_hog = dlib.get_frontal_face_detector()

# .detector_hog를 이용해 얼굴의 bounding box 추출
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)
```

```python
print(dlib_rects)

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_show, (l, t), (r, b), (0, 255, 0), 2, lineType=cv2.LINE_AA)
    
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 저장한 landmark 모델 불러오기
model_path = '/workspace/userdisk/AIFFEL_quest_rs/Exploration/Ex03/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

# 얼굴 영역 박스 마다 face landmark를 찾아냄
list_landmarks = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
```

```python
# landmark 출력
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 5, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
# 스티커 적용을 위해 코 위치 확인하기
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[30])
    x = landmark[30][0]
    y = landmark[30][1]
    w = h = dlib_rect.width()
    print (f'(x, y) : ({x}, {y})')
    print (f'(w, h) : ({w}, {h})')
```

```python
# 준비한 스티커 이미지를 읽어서 적용
sticker_path = os.getenv('HOME')+'/aiffel/AIFFEL_quest_rs/Exploration/Ex03/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w, h))
print(img_sticker.shape)
```

```python
# 원본 이미지에 스티커 이미지를 추가하기 위해 x, y 좌표 조정
sticker_h, sticker_w = img_sticker.shape[:2]

# 스티커 중앙이 코 좌표 (x, y)에 오도록 조정
refined_x = x - (sticker_w // 2)
refined_y = y - (sticker_h // 2)
print (f'(x, y) : ({refined_x}, {refined_y})')
```

```python
# 좌표가 음수로 나올 경우 crop
if refined_x < 0:
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0

if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0
    
print(f'(x,y) : ({refined_x},{refined_y})')
```

```python
# 원본의 스티커가 검은색에 투명한 배경이라서 색 자체를 인식하지 못한 것 같다.
# 그러다보니 색을 없애려고 하면 자꾸 검은 부분이 사라지고 오히려 배경 부분만 흰색이 됐다.
# 그래서 Chatgpt에게 물어봤더니 알파 채널을 쓰는 등 다른 배우지 않은 복잡한 방법을 소개해줬다.
# 그렇게 여러 번 프롬프트한 결과 흰색 배경을 만들어서 처리하는 방법으로 수정했다.

# 흰색 배경 마스크 만들기 (흰색이면 True)
white_mask = np.all(img_sticker == [255, 255, 255], axis=-1)

# 스티커 붙일 위치에서 원본 이미지 영역 추출
sticker_area = img_show[refined_y:refined_y + img_sticker.shape[0],
                       refined_x:refined_x + img_sticker.shape[1]]

# 마스크를 3채널로 확장 (스티커와 원본 이미지가 모두 3채널이기에 shape를 맞춰줘야 함)
white_mask_3ch = np.stack([white_mask]*3, axis=-1)

# 흰색인 부분은 원본을 유지, 나머지는 스티커 적용
img_show[refined_y:refined_y + img_sticker.shape[0],
        refined_x:refined_x + img_sticker.shape[1]] = \
    np.where(white_mask_3ch, sticker_area, img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 얼굴 검출 및 랜드마크 없는 원본 이미지에 스티커 적용

white_mask = np.all(img_sticker == [255, 255, 255], axis=-1)

sticker_area = img_bgr[refined_y:refined_y + img_sticker.shape[0],
                       refined_x:refined_x + img_sticker.shape[1]]

white_mask_3ch = np.stack([white_mask]*3, axis=-1)

img_bgr[refined_y:refined_y + img_sticker.shape[0],
        refined_x:refined_x + img_sticker.shape[1]] = \
    np.where(white_mask_3ch, sticker_area, img_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 사진에 눈, 코, 입이 잘 나와있고 정면이라 그런지 얼굴 검출이 잘 됐다.
# 만약 옆에 낙서에 코랑 눈썹이 있었다면 낙서들도 얼굴 검출이 됐을지 궁금하다.
# 매번 BGR2RGB를 해야되는지, 한 번만 적용하고 재사용할 순 없는지 궁금하다.
# 프롬프트 엔지니어링에 대해서도 공부를 더 해야겠다.
```

```python
for dlib_rect, landmark in zip(dlib_rects, list2_landmarks):
    print(landmark[30])
    x = landmark[30][0]
    y = landmark[30][1]
    w = h = dlib_rect.width()
    print (f'(x, y) : ({x}, {y})')
    print (f'(w, h) : ({w}, {h})')

img2_sticker = cv2.imread(sticker_path)
img2_sticker = cv2.resize(img2_sticker, (w, h))
print(img2_sticker.shape)

sticker2_h, sticker2_w = img2_sticker.shape[:2]

refined_x = x - (sticker2_w // 2)
refined_y = y - (sticker2_h // 2)
print (f'(x, y) : ({refined_x}, {refined_y})')

white_mask = np.all(img2_sticker == [255, 255, 255], axis=-1)

sticker2_area = img2_bgr[refined_y:refined_y + img2_sticker.shape[0],
                       refined_x:refined_x + img2_sticker.shape[1]]

white_mask_3ch = np.stack([white_mask]*3, axis=-1)

img2_bgr[refined_y:refined_y + img2_sticker.shape[0],
        refined_x:refined_x + img2_sticker.shape[1]] = \
    np.where(white_mask_3ch, sticker2_area, img2_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
