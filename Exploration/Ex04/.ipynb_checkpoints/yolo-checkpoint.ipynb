```python
# 1. ultralytics 설치
!pip install ultralytics

# 2. YOLO 불러오기
from ultralytics import YOLO

# 3. 모델 로드 (YOLOv8 Segmentation용)
model = YOLO("yolov8s-seg.pt")  # 또는 yolov8s-seg.pt 등으로 바꿔도 됨

# 결과 예측
results = model.predict(source="image2.png", task='segment')

# 이미지 + 마스크 시각화 (matplotlib 사용)
import matplotlib.pyplot as plt
import cv2

# 결과에서 첫 번째 이미지 가져오기
seg_result = results[0]
img_with_mask = seg_result.plot()  # 마스크 + 박스 등 그려진 이미지

# OpenCV는 BGR이므로 RGB로 바꿔서 출력
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(img_with_mask, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("YOLO Segmentation Result")
plt.show()
```

```python
# YOLOv11 불러오기
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# 모델 로드 (YOLOv11 Nano 버전, 필요시 's', 'm', 'l', 'x'로 변경 가능)
model = YOLO("yolo11n-seg.pt")  # 또는 yolo11s-seg.pt 등

# 이미지 파일 경로
img_path = "image2.png"

# 세그멘테이션 예측 수행
results = model(img_path)

# 결과 이미지 얻기 (마스크, 바운딩 박스 포함)
seg_result = results[0]
img_with_mask = seg_result.plot()

# 시각화 (cv2.imshow 대신 matplotlib 사용)
plt.figure(figsize=(8, 8))
plt.imshow(cv2.cvtColor(img_with_mask, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("YOLOv11 Segmentation Result")
plt.show()
```
